```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Read the CSV file
file_path = 'bbc-news-summary.csv'  # Update with the path to your CSV file
data = pd.read_csv(file_path)[:100]


# Step 2: Extract the 'Summaries' column
summaries = data['Summaries']

# Step 3: Count the number of statements in each summary
# Assuming each statement ends with a period (.), you can split by '.' and count
statement_counts = summaries.apply(lambda x: len(x.split('.')) - 1 if isinstance(x, str) else 0)

# Step 4: Plot the frequency distribution of the number of statements
plt.figure(figsize=(20, 20))
sns.histplot(statement_counts, bins=range(1, max(statement_counts)+2), kde=False)
plt.xlabel('Number of Statements')
plt.ylabel('Frequency')
plt.title('Frequency Distribution of Number of Statements in Summaries')
plt.xticks(range(1, max(statement_counts)+1))
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Read the CSV file
file_path = 'bbc-news-summary.csv'  # Update with the path to your CSV file
data = pd.read_csv(file_path)

# Step 2: Extract the 'Articles' and 'Summaries' columns
articles = data['Articles']
summaries = data['Summaries']

# Step 3: Count the number of statements in each article and summary
# Assuming each statement ends with a period (.), you can split by '.' and count
article_statement_counts = articles.apply(lambda x: len(x.split('.')) - 1 if isinstance(x, str) else 0)
summary_statement_counts = summaries.apply(lambda x: len(x.split('.')) - 1 if isinstance(x, str) else 0)

# Step 4: Create a DataFrame with the counts
statement_counts = pd.DataFrame({
    'Article Statements': article_statement_counts,
    'Summary Statements': summary_statement_counts
})

# Step 5: Plot the data
plt.figure(figsize=(10, 6))
sns.scatterplot(data=statement_counts, x='Article Statements', y='Summary Statements')
plt.xlabel('Number of Statements in Articles')
plt.ylabel('Number of Statements in Summaries')
plt.title('Number of Statements in Articles vs. Summaries')
plt.show()
```
